### Analysis
    The Capomulin and Ramicane regimens exhibited the smallest average tumor volumes, along with lower variance. It's important to note that these two drug regimens also had a higher number of time points compared to others, potentially introducing bias to the average tumor volume values. Additionally, an equal distribution of male and female mice, approximately 50% each, was ensured. Furthermore, there appears to be a correlation between tumor volume and mouse weight, particularly within the Capomulin regimen.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df=pd.merge(mouse_metadata,study_results,on="Mouse ID")

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
combined_df["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicatemice = combined_df[combined_df.duplicated(['Mouse ID','Timepoint'])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicatemice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df=combined_df.drop_duplicates(subset=["Mouse ID","Timepoint"])
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
df3=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].describe()
df_summary_per_regimen=pd.DataFrame(df3[['mean','50%']])
df_summary_per_regimen['variance']=df3['std']**2
df_summary_per_regimen['standard deviation']=df3['std']
df_summary_per_regimen['SEM']=df3['std']/df3['count']
df_summary_per_regimen.rename(columns={'50%':'median'})
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
df_summary_method_aggregate=clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate(['mean','median','var','std','sem'])
df_summary_method_aggregate
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df5=clean_df.groupby('Drug Regimen')['Timepoint'].max()
df5

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
maxtimepoint=max(clean_df['Timepoint'])
maxtimepoint

df_filter4drugs=clean_df.loc[(clean_df['Drug Regimen']=='Capomulin')|(clean_df['Drug Regimen']=='Ramicane')|
                        (clean_df['Drug Regimen']=='Infubinol')|(clean_df['Drug Regimen']=='Ceftamin'),:]
df_filter4drugs_Maxtime=df_filter4drugs.loc[df_filter4drugs['Timepoint']==maxtimepoint]
df_filter4drugs_Maxtime
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments=[x for x in df_filter4drugs_Maxtime['Drug Regimen']]
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers.   
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
# Determine outliers using upper and lower bounds
quartiles=df_filter4drugs_Maxtime['Tumor Volume (mm3)'].quantile([.25,.5,.75])
lowerq=quartiles[.25]
upperq=quartiles[0.75]
iqr=upperq - lowerq
lower_bound=lowerq - (1.5*iqr)
upper_bound=upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
outliers_df=df_filter4drugs_Maxtime.loc[(df_filter4drugs_Maxtime['Tumor Volume (mm3)']<lower_bound)|
                                        (df_filter4drugs_Maxtime['Tumor Volume (mm3)']>upper_bound),:]
print(outliers_df)
print('Empty data frame means no outliers.')

fig1, ax1 = plt.subplots()
ax1.set_title('Tumor Volume (mm3) for the 4 drugs')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.boxplot(df_filter4drugs_Maxtime['Tumor Volume (mm3)'])
plt.show()
print('There is no visible outlier.')
print('CONCLUSION: Both methods show no outliers when considering tumour volumes for all 4 drugs combined.')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
a=0
b=0
fig2, ax1 = plt.subplots(2,2)
for drug in ['Capomulin','Ramicane','Infubinol','Ceftamin']:
    ser_eachdrug=df_filter4drugs_Maxtime.loc[df_filter4drugs_Maxtime['Drug Regimen']=="Capomulin",'Tumor Volume (mm3)']
    ax1[a,b].set_title(drug)
    ax1[a,b].set_ylabel('Tumor Volume (mm3)')
    ax1[a,b].boxplot(ser_eachdrug)
    if b==0:
        b=1
    elif a==0 and b==1:
        a=1
        b=0
plt.tight_layout()
plt.show()
print('CONCLUSION: No outliers for tumour volumes when considering each drug one by one.')
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
random_mouse_Capomulin=clean_df.loc[clean_df['Drug Regimen']=='Capomulin'].sample(n=1)['Mouse ID']
r = random_mouse_Capomulin.astype('string').values[0]
df_for_line_plot=clean_df.loc[clean_df['Mouse ID'] == r,:]
plot1=df_for_line_plot.plot(x="Timepoint",y="Tumor Volume (mm3)", marker='o',
                      title="Tumor volume vs time point for a mouse treated with Capomulin")
plot1.set_ylabel(r'$Tumor~Volume~(mm^3)$')
plt.savefig("./figures/Line plot of tumour volume vs time point_mouse_Capomulin.png")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
df_Capomulin=clean_df.loc[clean_df['Drug Regimen']=='Capomulin',['Mouse ID','Weight (g)','Tumor Volume (mm3)']]

df_average_Capomulin_4plot=df_Capomulin.groupby('Mouse ID').mean()

plt.scatter(df_average_Capomulin_4plot['Weight (g)'],df_average_Capomulin_4plot['Tumor Volume (mm3)'])
plt.title("Average tumor volume vs mouse weight for Capomulin regimen")
plt.xlabel("Average mouse weight")
plt.ylabel("Average Tumor volume")
plt.tight_layout()
plt.savefig("./figures/TumorVol vs Mouse weight for Capomulin.png")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from scipy.stats import linregress

mouse_weight=df_average_Capomulin_4plot['Weight (g)']
tumour_vol=df_average_Capomulin_4plot['Tumor Volume (mm3)']

print(f"Correlation coefficient for mouse weight and average tumour volume for Capomulin regime is \
{st.pearsonr(mouse_weight,tumour_vol)[0]}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, tumour_vol)
regress_values = mouse_weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(mouse_weight,tumour_vol)
plt.plot(mouse_weight,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumour Volume (mm3)')
plt.show()
```
